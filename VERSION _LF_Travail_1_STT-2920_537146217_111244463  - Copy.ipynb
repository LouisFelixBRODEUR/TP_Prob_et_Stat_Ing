Nous allons ici vérifiez ce fait empiriquement à l'aide d'une simulation basée sur $m=10000$ réplications, en suivant la procédure suivante. Nous allons faire le cas $n=5.$

*   Simuler des variables $X_1,X_2,\ldots,X_5$ telles que

    $X_1\sim\mathrm{exponentielle}(3),$

    $X_2\sim\mathrm{exponentielle}(4),$

    $X_3\sim\mathrm{exponentielle}(5),$

    $X_4\sim\mathrm{exponentielle}(6),$

    $X_5\sim\mathrm{exponentielle}(7).$    

*   Prendre le minimum $Q$ des 5 valeurs obtenues.
*   Répéter les deux étapes précédentes $m$ fois et stocker ces $m$ valeurs de $Q.$

```python
#Ne pas changer cette cellule
import numpy as np
from math import*
import matplotlib.pyplot as plt
import statistics
from scipy.stats import expon
from scipy.stats import gamma
#######
```

Modifier le code ci-dessous.
Vous aurez besoin de consulter la page
https://numpy.org/doc/2.1/reference/random/generated/numpy.random.exponential.html
et le Notebook présenté en classe dans le chapitre 8 où l'on a fait des simulations.

```python
#Définition des paramètres de départ.
#Nombre de simulations (ne pas modifier)
m=10000
#Définition des paramètres des lois exponentielles (ne pas modifier)
lamb1=3
lamb2=4
lamb3=5
lamb4=6
lamb5=7
##########

#Préparation du tableau (une seule ligne de m valeurs) Q qui sera à remplir (ne pas modifier)
Q=np.zeros(m)

#Boucle pour la simulation

for i in range(m):
  #Changer les 5 prochaines lignes pour que chacun des xi contient un nombre tiré de leur loi exponentielle respective. Vous utiliserez la fonction np.random.exponential().
  #Attention, cette fonction prend en entrée un des paramètres qui n'est pas directement lambda mais qui dépend de lui, bien lire la page de référence. Votre code devrait avoir une seule ligne par xi.
  x1=np.random.exponential(1/lamb1)
  x2=np.random.exponential(1/lamb2)
  x3=np.random.exponential(1/lamb3)
  x4=np.random.exponential(1/lamb4)
  x5=np.random.exponential(1/lamb5)

  #On emmagasine la valeur du minimum des xi dans Q. (Ne pas modifié)
  Q[i]=min(x1,x2,x3,x4,x5)

#Fin de la boucle pour la simulation
```

```python
#Aperçu des valeurs de Q obtenus.
print(Q)
```

```python
#Définissons lambda_star le paramètre de la loi exponentielle de la loi de Q, en fonction des lambda des lois des X_i
#Une ligne à modifier (cette ligne ne comportera aucun chiffre, mais le nom de paramètres définis plus haut apparaîtra dans un calcul)
lamb_star=lamb1 + lamb2 + lamb3 + lamb4 + lamb5
```

**Question 2**

Soient $X_1,X_2,\ldots, X_n$ i.i.d et tels que $\mathbb{E}[X_i]=\mu$ et que $\mathrm{Var}[X_i]=\sigma^2,$

Nous avons vu en classe que l'estimateur variance échantillonnale corrigée $$S^2=\dfrac{1}{n-1}\sum_{i=1}^n \left(X_i-\overline{X}\right)^2$$
est un estimateur sans biais de $\sigma^2,$ c'est-à-dire que $$\mathbb{E}[S^2]=\sigma^2$$

Dans cette question, nous allons vérifier empiriquement ce fait à l'aide d'une simulation et comparer avec l'estimateur similaire (mais non-corrigée):
$$S_{(n)}^2=\dfrac{1}{n}\sum_{i=1}^n\left(X_i-\overline{X}\right)^2$$

Nous allons utiliser une expérimentation basée sur $m=100\,000$ réplications de chacune de ces deux variables dans le cas où les $$X_i\sim\mathcal{N}(\mu;\sigma^2)$$ avec $\mu=15,$ $\sigma=3$ et $n=10.$

Remarque: La structure générale de la simultation est similaire à la question précédente, mais nous allons ici stocker les valeurs des m réplications de chacune des variables $X_i$ dans un tableau au lieu de faire une boucle for et de broyer les valeurs de l'itération précédente.

*  Générer $m$ fois les variables $X_1,X_2,...,X_n$ et stocker les valeurs dans un tableau n lignes et m colonnes.
*  Pour chaque réplication (chaque colonne), calculez la valeur de $S^2$ et la valeur de $S_{(n)}^2.$
* Comparer l'histogramme des valeurs de $S^2$ et celui des valeurs de $S_{(n)}^2.$
* Comparer les diagrammes en boites des valeurs de $S^2$ et celui des valeurs de $S_{(n)}^2.$
* Calculer la moyenne des $m$ valeurs de $S^2$ (cette moyenne estime $\mathbb{E}[S^2]$) et calculer la moyenne des $m$ valeurs de $S_{(n)}^2$ (cette moyenne estime $\mathbb{E}\left[S_{(n)}^2\right]$.   




Voir le notebook sur les simulations présentées au chapitre 8 au besoin.

```python
#À MODIFIER: les paramètres de départs ci-dessous (voir l'énoncé ci-dessus)
m=100000
n=10
mu=15
sigma=3
```

```python
#Ce tableau est de dimension n lignes et m colonnes et est rempli de 0 initialement
data=np.zeros((n, m)) #Ne pas modifier.

#On remplit le tableau, chaque ligne du tableau doit contenir m réalisations de la loi normale de moyenne et d'écart-type approprié
#Modifier la ligne de code dans la boucle for (utilisez la fonction np.random.normal()), attention de bien regarder ce que cette fonction prend en entrées.
for i in range(n):
  data[i,]=np.random.normal(mu, sigma, m)
#Fin de la boucle for.

#On calcule la variance échantillonnale corrigée (S^2) et la variance échantillonnale non-corrigée(S(n)^2) pour chacune des m réplications.
#La fonction np.apply_along_axis(fonction, 0, data) calcule sur chaque colonne du tableau data la valeur de la fonction donnée en premier paramètre.
#Ainsi, cela donne en sortie les m valeurs du calcul de fonction. La valeur pour la première colonne, la valeur pour la deuxième colonne,...,la valeur pour la m ième colonne.
#La valeur du paramètre 0  indique que la fonction est calculer par colonne (ne pas modifier ce paramètre).
#Vous devez donc modifier l'entrée np.mean ici par une fonction python qui calcul la variance corrigée et la fonction python qui calcul la variance non-corrigée.
#Voir le notebook sur la statistique descriptive au besoin.
variances_corrigees=np.apply_along_axis(lambda x: np.var(x, ddof=1), 0, data) #À modifier
variances_non_corrigees=np.apply_along_axis(lambda x: np.var(x, ddof=0), 0,data) #À modifier
```

```python
#Apperçu des valeurs de S^2 et de S_(n)^2 (ne pas modifier)
print(variances_corrigees)
print(variances_non_corrigees)
```

```python
#Rien à modifier dans cette cellule.
#Affichage de l'histogramme de m valeurs de S^2
plt.hist(variances_corrigees,bins='auto', density=True,edgecolor="black")
plt.xlabel('Valeurs de S^2')
plt.ylabel('Proportion ajustée')
plt.title("Histogramme des valeurs de S^2 obtenues\n basé sur {} simulation".format(m))
plt.show()

#Affichage de l'histogramme de m valeurs de S_(n)^2
plt.hist(variances_non_corrigees,bins='auto', density=True,edgecolor="black")
plt.xlabel('Valeurs de S(n)^2')
plt.ylabel('Proportion ajustée')
plt.title("Histogramme des valeurs de S_(n)^2 obtenues\n basé sur {} simulation".format(m))
plt.show()
```

```python
#Cellule à modifier.
#Faire le diagramme en boites des deux séries de données dans le même diagramme (voir le Notebook sur la statistique descriptive au besoin)
plt.boxplot([variances_corrigees, variances_non_corrigees], labels=['S^2', 'S(n)^2'])
plt.title("Diagramme en boites de S^2 et S(n)^2")
plt.ylabel("Valeurs")
plt.show()
```

```python
#Calculer la moyenne des deux séries de données en modifiant les deux lignes ci-dessous.
moyenne_var_corrigee=np.mean(variances_corrigees)
moyenne_var_non_corrigee=np.mean(variances_non_corrigees)

print("La moyenne des valeurs de S^2 est ",moyenne_var_corrigee)
print("La moyenne des valeurs de S_(n)^2 est", moyenne_var_non_corrigee)
```

Comparez les deux valeurs obtenus par rapport à la valeur que les deux estimateurs estiment.

Commentaire de l'étudiant:
Les deux valeurs obtenues sont proches l’une de l’autre, mais la moyenne de 8.98 est légèrement plus élevée que celle de 8.08. Cela correspond exactement à ce que l’on s’attend à observer : S^2 est un estimateur corrigé, donc sans biais, de la variance théorique, tandis que S_(n)^2 est un estimateur non corrigé, qui a tendance à sous-estimer la variance réelle. Vu la forme de l'histogramme, il est plausible que la loi gamma soit un bon modèle pour décrire la variable aléatoire $G$ qui a produit ces données.
On croit donc que $$G\sim\mathrm{gamma}(\alpha,\lambda)$$
On souhaite donc trouver, parmi toutes les lois gamma, celle dont la densité colle le mieux à l'histogramme des valeurs observées de $G.$ On doit donc choisir le meilleur $\alpha$ et le meilleur $\lambda$ à l'aide des données échantillonnale observées. On va donc estimer $\alpha$ et $\lambda$ à l'aide de nos données.

On sait si $G\sim\mathrm{gamma}(\alpha,\lambda),$ alors les formules de l'espérance et de la variance de la loi gamma nous donne que

$$\mathbb{E}[G]=\dfrac{\alpha}{\lambda}\qquad\text{ et }\qquad \mathrm{Var}[G]=\dfrac{\alpha}{\lambda^2}.$$

À l'aide de nos données échantillonnales, on peut estimer $\mathbb{E}[G]$ par $\overline{g},$ la moyenne échantillonnale et $\mathrm{Var}[G]$ par $s^2,$ la variance corrigée échantillonnale. on a donc que

$$\overline{g}\approx\dfrac{\alpha}{\lambda}\qquad\text{ et }\qquad s^2\approx\dfrac{\alpha}{\lambda^2}.$$

Ainsi, si on isole $\alpha$ et $\lambda$ en fonction de $\overline{g}$ et $s^2$ dans le système de deux équations, on obtiendra les formules à utiliser pour estimer $\alpha$ et $\lambda$

Obtenez un intervalle de confiance au niveau de confiance 99.5% pour résistance moyenne théorique des composantes produites par ce procédé et donnez une interprétation contextuelle de cette intervalle de confiance. Voir le notebook sur l'inférence statistique présenté dans le chapitre 13.

```python
#À FAIRE
alpha = 0.005  # pour un IC à 99.5%
moy = np.mean(R)
s = np.std(R, ddof=1)
n = len(R)

tcrit = stats.t.ppf(1 - alpha/2, df=n-1)

borne_inf = moy - tcrit * s / np.sqrt(n)
borne_sup = moy + tcrit * s / np.sqrt(n)

print("L'intervalle de confiance à 99.5% est : [", borne_inf, ",", borne_sup, "]")
```

Interprétation contextuelle de l'intervalle de confiance:

L’intervalle de confiance à 99.5% obtenu, soit $[99.915;100.485]$ micro-ohms, représente l’ensemble des valeurs plausibles pour la vraie résistance moyenne du procédé. Comme cet intervalle inclut la valeur critique de 100 micro-ohms, il n’est pas possible d’affirmer avec un niveau de confiance aussi élevé que la résistance moyenne dépasse cette limite. Autrement dit, même si certaines observations sont légèrement au-dessus de 100 micro-ohms, l’incertitude statistique est suffisamment grande pour que la moyenne réelle puisse être égale ou légèrement inférieure à 100 micro-ohms.
